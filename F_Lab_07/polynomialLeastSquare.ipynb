```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data points
a = np.array([1, 3, 5, 7, 9], dtype=int)
b = np.array([5.7, 9.3, 15.1, 20.4, 32.8], dtype=float)

# Degree of the polynomial to fit
n = 2
m = len(a)

# step-1
a_sum = [np.sum(a**k) for k in range(2 * n + 1)]
ab_sum = [np.sum((a**k)*b) for k in range(n+1)]

# step-2
X = np.zeros((n+1, n+1))
Y = np.zeros(n+1)

# step-3
for i in range(n+1):
    for j in range(n+1):
        X[i, j] = a_sum[i+j]
    Y[i] = ab_sum[i]


c = np.linalg.solve(X, Y)
print("Polynomial coefficients (c0, c1, c2) : ", c)

# step-4

b_pred = np.zeros_like(a, dtype=float)
for i in range(n + 1):
    b_pred += c[i] * (a ** i)


# plot

plt.scatter(a, b, color="red", label="Data Points")
plt.plot(a, b_pred, color="darkblue", label=f"Degree{n} fit")
plt.xlabel("a")
plt.ylabel("b")
plt.grid()
plt.legend()
plt.title("Polynomial Least Squares Fit")
plt.show()
```
